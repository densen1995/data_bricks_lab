## Exploratory data analysis for simulated event

```python
from pyspark.sql import functions as F



RAW_PATH = "/Volumes/marathos/default/raw/simulated/2026-06-03_201652_northern_lights_ultra"

```

```python
df = (
    spark.read
    .option("header", True)
    .csv(RAW_PATH)
)

df.limit(10).display()
```

## Shape

```python
n_rows = df.count()
n_cols = len(df.columns)
print(f"rows: {n_rows:,}")
print(f"columns: {n_cols}")
```

## Country representation

```python
df.groupBy("Athlete country") \
    .count() \
    .orderBy(F.desc("count")) \
    .limit(25) \
    .display()
```
